```python
%%time
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time

quote_page = requests.get('https://www.wtatennis.com/calendar')
soup = BeautifulSoup(quote_page.content,'html.parser')

header = ['tournament','date','location','prize','category','link']
data = []

for table in soup.find_all("tr", attrs={"class":"premier"}):
    for tournament in table.find_all("td", attrs={"data-th":"tournament"}):
        tournament = tournament.text
    for date in table.find_all("td", attrs={"data-th":"Location"}):
        date = date.text
    for location in table.find_all("td", attrs={"data-th":"location"}):
        location = location.text
        
    for prize in table.find_all("td", attrs={"data-th":"prize"}):
        prize = prize.text
    for category in table.find_all("td", attrs={"data-th":"category"}):
        category = category.text
    links = ("https://www.wtatennis.com" + table.a.get('href'))
    print(tournament)
    print(date)
    print(location)
    print(prize)
    print(category)
    print(links)
    data.append((tournament,date,location,prize,category,links))

df = pd.DataFrame(data,
    columns = header
)
df.to_csv('PS#2_wtacalendar2019.csv', sep='\t', encoding='utf-8')
```
